**Spark Streamining**

```python
entities=['customers','vehicles','drivers','locations','payments','trips']
```

```python
for entity in entities:

    df_batch=spark.read.format("csv").option("header","true").option("inferSchema","true").load(f"/Volumes/uber_cata/uber_schema/source/{entity}/")

    schema_entity=df_batch.schema

    df=spark.readStream.format("CSV").option("header",True).schema(schema_entity).load(f"/Volumes/uber_cata/uber_schema/source/{entity}/")

    df.writeStream.format("delta").outputMode("append").option("checkpointLocation",f"/Volumes/uber_cata/bronze/checkpoint/{entity}").trigger(once=True).toTable(f"uber_cata.bronze.{entity}")
```
